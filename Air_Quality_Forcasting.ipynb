**Forecasting with FB Prophet Algorithm**


```python
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y') # Use format day/month/year
print(date_info)
```

```python
time_info = air_quality_data['Time']
print(time_info)
```

```python
time_info = time_info.apply(lambda x:x.replace('.', ':'))
print(time_info)
```

```python
print(type(time_info))
print(type(date_info))
```

```python
from datetime import datetime
date_time = pd.concat([date_info, time_info], axis=1)
print(date_time)
```

```python
date_time.shape
```

```python
# combining date and time

date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)

```

```python
date_time.head()
```

```python
date_time.info()
```

we  have converted ds from object datatype to date time format

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y'] = air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
# tranning the prophet model

model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
